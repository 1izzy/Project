```python
df_kakao=pd.read_csv('카카오페이_all.csv')
df_naver=pd.read_csv('네이버페이_all.csv')
df_samsung=pd.read_csv('삼성페이_all.csv')
```

```python
def 전처리(dataframe) :
    df=dataframe[['content','score','at']]
    df.sort_values(['at'],ascending=False,inplace=True) #시간대로 정렬
    df.reset_index(drop=True,inplace=True) #인덱스 재설정
#     df['at']=df['at'].str.split(" ").str.get(0) #날짜만 추출
    return df
```

```python
df_n=전처리(df_naver)
df_k=전처리(df_kakao)
df_s=전처리(df_samsung)
```

```python
df_n=df_n[['at','score','content']]
df_n.columns=['DATE','STAR','REVIEW']
df_k=df_k[['at','score','content']]
df_k.columns=['DATE','STAR','REVIEW']
df_s=df_s[['at','score','content']]
df_s.columns=['DATE','STAR','REVIEW']
```

```python
df_n.to_csv('네이버페이_플레이스토어.csv',index=False,encoding='utf-8-sig')
df_k.to_csv('카카오페이_플레이스토어.csv',index=False,encoding='utf-8-sig')
df_s.to_csv('삼성페이_플레이스토어.csv',index=False,encoding='utf-8-sig')
```
