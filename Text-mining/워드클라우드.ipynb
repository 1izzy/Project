```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
df=pd.read_csv('별점별 _ 빈도분석.csv')
```

```python
naver=df[df['어플']=='네이버']
samsung=df[df['어플']=='삼성']
kakao=df[df['어플']=='카카오']
```

# 네이버 워드클라우드

```python
from matplotlib import cm 
from matplotlib.colors import ListedColormap
color_modified = cm.get_cmap('YlGn', 256)
newcmp = ListedColormap(color_modified(np.linspace(0.5, 1, 256)))
newcmp
```

```python
for i in range(5):
    naver_df=naver[naver['별점']==i+1]
    dicts=naver_df[['단어','빈도']].set_index('단어').to_dict()
    dicts=dicts.get('빈도')
    #워드클라우드 코드 
    wordcloud = WordCloud(font_path='C:\windows/fonts/HMKMRHD.ttf', background_color = 'white',random_state=12,max_words=100,colormap=newcmp,
                      width = 1000, height = 800).generate_from_frequencies(dicts)
    plt.figure(figsize=(10,10))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    wordcloud.to_file('네이버별점{}_wc.png'.format(i+1))
```
