```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
naver=pd.read_csv('네이버페이_클러스터링.csv')
kakao=pd.read_csv('카카오페이_클러스터링.csv')
samsung=pd.read_csv('삼성페이_클러스터링.csv')
```

```python
dfnaver=naver[(naver.result!=0) &( naver.result != -1)]
dfkakao=kakao[(kakao.result!=0) &( kakao.result != -1)]
dfsamsung=samsung[(samsung.result!=0) &( samsung.result != -1)]
dfnaver.reset_index(drop=True,inplace=True)
dfsamsung.reset_index(drop=True,inplace=True)
dfkakao.reset_index(drop=True,inplace=True)
```

```python
print("네이버 클러스터 수 :",len(naver.result.unique())-2)
print("카카오 클러스터 수 :",len(kakao.result.unique())-2)
print("삼성 클러스터 수 :",len(samsung.result.unique())-2)
```

```python
from collections import Counter 
from sklearn.feature_extraction.text import TfidfVectorizer
keyword_naver=[]
corpus=[]
for i in range(11):
    data=dfnaver[dfnaver.result== i+1]
    data.token_REVIEW=data.token_REVIEW.apply(lambda x: x.split(' '))
    word_list=sum(data['token_REVIEW'].to_list(),[])
    word=" ".join(word_list)    
#     c = Counter(word_list)
#     keyword_naver.append(c.most_common(1))
    corpus.append(word)
TfidfV=TfidfVectorizer()
tfidfv = TfidfV.fit_transform(corpus)
tfidf_df = pd.DataFrame(tfidfv.toarray(), columns = TfidfV.get_feature_names())
# naverdf=pd.DataFrame()
# for j in range(11):
#     data=dfnaver[dfnaver['result']==j+1]
#     data['키워드']=keyword_naver[j][0][0]
#     naverdf=naverdf.append(data)   
```

```python
keyword_kakao=[]
for i in range(61):
    data=dfkakao[dfkakao.result== i+1]
    data.token_REVIEW=data.token_REVIEW.apply(lambda x: x.split(' '))
    word_list=sum(data['token_REVIEW'].to_list(),[])
    c = Counter(word_list)
    keyword_kakao.append(c.most_common(1))
kakaodf=pd.DataFrame()
for i in range(61):
    data=dfkakao[dfkakao['result']==i+1]
    data['키워드']=keyword_kakao[i][0][0]
    kakaodf=kakaodf.append(data)   
```

```python
keyword_samsung=[]
for i in range(406):
    data=dfsamsung[dfsamsung.result== i+1]
    data.token_REVIEW=data.token_REVIEW.apply(lambda x: x.split(' '))
    word_list=sum(data['token_REVIEW'].to_list(),[])
    c = Counter(word_list)
    keyword_samsung.append(c.most_common(1))
samsungdf=pd.DataFrame()
for i in range(406):
    data=dfsamsung[dfsamsung['result']==i+1]
    data['키워드']=keyword_samsung[i][0][0]
    samsungdf=samsungdf.append(data)   
```
